```python
import pandas as pd
import numpy as np
filename = 'Survey On Workload and Mental Health in University Students.csv'
df = pd.read_csv(filename)

print("--- Original Data Info ---")
df.info()
new_names = {
    "You're Currently On": "year_study",
    "Gender": "gender",
    "Age": "age",
    "CGPA": "cgpa",
    "How many hours do you spend on studying per week?": "study_hours",
    " How many courses are you currently enrolled in?": "courses", 
    "Do you participate in extracurricular activities?": "extra_activities",
    "If yes, how many hours per week?": "extra_hours",
    "Do you have a part-time job?": "job",
    "If yes, how many hours per week?.1": "job_hours",
    "On a scale of 1–5, how would you rate your stress level this semester? (1 = Very Low, 5 = Very High)": "stress_level",
    "On a scale of 1–5, how would you rate your anxiety level? (1 = No anxiety, 5 = Extreme anxiety)": "anxiety_level",
    "On average, how many hours of sleep do you get per night?": "sleep_hours",
    "On a scale of 1–10, how would you rate your overall sleep quality?": "sleep_quality"
}

df = df.rename(columns=new_names)


df = df.drop(columns=['Timestamp'], errors='ignore')
df['extra_hours'] = df['extra_hours'].fillna(0)
df['job_hours'] = df['job_hours'].fillna(0)

print("\nMissing values after filling hours:\n", df.isnull().sum())
binary_map = {'Yes': 1, 'No': 0}
df['extra_activities_enc'] = df['extra_activities'].map(binary_map)
df['job_enc'] = df['job'].map(binary_map)


gender_map = {'Female': 0, 'Male': 1}
df['gender_enc'] = df['gender'].map(gender_map)

df['year_study_cleaned'] = df['year_study'].astype(str).str.replace(r'\D+', '', regex=True)
df['year_study_enc'] = pd.to_numeric(df['year_study_cleaned'])

final_columns = [
    'year_study_enc', 'gender_enc', 'age', 'cgpa',
    'study_hours', 'courses',
    'extra_activities_enc', 'extra_hours',
    'job_enc', 'job_hours',
    'stress_level', 'anxiety_level',
    'sleep_hours', 'sleep_quality'
]

df_clean = df[final_columns]

print("\n--- Final Cleaned Data Head ---")
print(df_clean.head())

print("\n--- Final Data Types ---")
print(df_clean.info())

df_clean.to_csv('cleaned_data.csv', index=False)
print("\nSuccess! 'cleaned_data.csv' has been saved.")
```
